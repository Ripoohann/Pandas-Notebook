```python
import numpy as np
import pandas as pd
```

```python
df  = pd.read_csv('C:/covid.csv')    # Importing a CSV file in 'df' Data Frame
```

```python
df.head(5)    # Prints the first (n) Number of values
```

```python
df.drop(['Active'], axis=1, inplace=True)    # Deletes the specified column    ( inplace = Trues :- makes changes in the 
                                             #                                   existing Data Frame does not create a copy and then show it)
```

```python
df
```

```python
df.rename(columns={"Country/Region" : "Region"}, inplace = True)    # Renames the specified Column
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df["Date"])  # Imp    # Converts the given string into default Pandas DatetimeIndex format
```

```python
df
```

```python
df.describe()    # Describes the data (mean, count, min, etc)
```

```python
Regions = df3['Region'].unique()
len(Regions)
```

```python
for idx in range(0, len(Regions)):
    C = df3[df3['Region']==Regions[idx]].reset_index()
    plt.scatter(np.arange(0,len(C)), C['Confirmed'], color = 'b', label = 'Confirmed', linewidth = 1)
    plt.scatter(np.arange(0,len(C)), C['Recovered'], color = 'g', label = 'Recovered', linewidth = 1)
    plt.scatter(np.arange(0,len(C)), C['Deaths'], color = 'r', label = 'Deaths', linewidth = 2)
    plt.title(Regions[idx])
    plt.xlabel('Days since first suspect')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.show()
    
```
